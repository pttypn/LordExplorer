```python
import re

def addlatlon(dataframe):
    newdf = dataframe.copy()
    #---------add lattitude and longitude columns
    latlon = re.compile('(-71\.\d+),(41\.\d+)')
    givelat = lambda LatLon: latlon.search(str(LatLon))[1] if isinstance(
                            latlon.search(str(LatLon)),re.Match) else None
    givelon = lambda LatLon: latlon.search(str(LatLon))[2] if isinstance(latlon.search(str(LatLon)),re.Match) else None

    newdf['Lattitude'] = newdf.latlon.apply(givelat)
    newdf['Longitude'] = newdf.latlon.apply(givelon)
    return newdf
```
